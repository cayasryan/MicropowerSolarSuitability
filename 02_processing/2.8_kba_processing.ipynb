```python
import geopandas as gpd

gdf = gpd.read_file("../00_data/Philippines_KBA.shp")
gdf
```

```python
centroid = gdf.geometry.centroid.iloc[0]  # Gets the centroid of the first polygon
print((centroid.x, centroid.y))
```

```python
file_path = "../01_processed_data/philippines_kba.geojson"

gdf = gdf.to_crs(epsg=32651)

# Save only the geometry to a GeoJSON file
gdf[['SitRecID','NatName','geometry']].to_file(file_path, driver="GeoJSON")
```
