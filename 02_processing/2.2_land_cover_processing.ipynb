```python
import geopandas as gpd
import fiona

# Path to your GDB file
gdb_path = "../00_data/land_cover/LCM2020_VISAYAS_PH_UTMZ51_PRS92.gdb"

# List all available layers in the GDB
layers = fiona.listlayers(gdb_path)
print("Layers:", layers)

with fiona.Env(GDAL_EXTRA_OPTIONS={'OGR_ORGANIZE_POLYGONS': 'SKIP'}):
    gdf_visayas = gpd.read_file(gdb_path, layer=layers[0])

gdf_visayas
```

```python
# Path to your GDB file
gdb_path = "../00_data/land_cover/LCM2020_MINDANAO_PH_UTMZ51_PRS92.gdb"

# List all available layers in the GDB
layers = fiona.listlayers(gdb_path)
print("Layers:", layers)

with fiona.Env(GDAL_EXTRA_OPTIONS={'OGR_ORGANIZE_POLYGONS': 'SKIP'}):
    gdf_mindanao = gpd.read_file(gdb_path, layer=layers[0])

gdf_mindanao
```

```python
# Path to your GDB file
gdb_path = "../00_data/land_cover/LCM2020_LUZON_PH_UTMZ51_PRS92.gdb"

# List all available layers in the GDB
layers = fiona.listlayers(gdb_path)
print("Layers:", layers)

with fiona.Env(GDAL_EXTRA_OPTIONS={'OGR_ORGANIZE_POLYGONS': 'SKIP'}):
    gdf_luzon = gpd.read_file(gdb_path, layer=layers[0])

gdf_luzon
```

```python
import pandas as pd

# Merge the three GeoDataFrames
gdf = gpd.GeoDataFrame(pd.concat([gdf_luzon, gdf_visayas, gdf_mindanao], ignore_index=True))

unique_class_labels = gdf[["CLASS_ID", "CLASS_NAME"]].drop_duplicates()

# Filter gdf
filtered_gdf = gdf[["CLASS_ID", "CLASS_NAME", "geometry"]]

unique_class_labels
```

## Merge Classes for Simplified Analysis

01: Terrestrial Forest: Open Forest (4), Closed Forest (1) \
02: Crop Areas: Annual Crop (16), Perennial Crop (17,10) \
03: Barren/Flatland: Brush/Shrubs (10), Open/Barren (13), Grassland (14) \
04: Built-up: Built-up (20) \
05: Wetlands & Water Bodies: Mangrove Forest (7), Inland Water (21), Fishpond (19), Marshland/Swamp (15)

```python
# Mapping old class IDs to new class IDs
class_mapping = {
    4: "1",  # Open Forest â†’ Terrestrial Forest
    1: "1",  # Closed Forest â†’ Terrestrial Forest
    16: "2", # Annual Crop â†’ Crop Areas
    17: "2", # Perennial Crop â†’ Crop Areas
    10: None, # Needs manual correction (Brush/Shrubs OR Perennial Crop)
    13: "3", # Open/Barren â†’ Barren/Flatland
    14: "3", # Grassland â†’ Barren/Flatland
    20: "4", # Built-up â†’ Built-up
    7:  "5", # Mangrove Forest â†’ Wetlands & Water Bodies
    21: "5", # Inland Water â†’ Wetlands & Water Bodies
    19: "5", # Fishpond â†’ Wetlands & Water Bodies
    15: "5"  # Marshland/Swamp â†’ Wetlands & Water Bodies
}

# Apply mapping, but handle CLASS_ID = 10 separately
filtered_gdf["NEW_CLASS_ID"] = filtered_gdf["CLASS_ID"].map(class_mapping)

# Manually correct misclassified CLASS_ID = 10 based on CLASS_NAME
filtered_gdf.loc[(gdf["CLASS_ID"] == 10) & (filtered_gdf["CLASS_NAME"].str.contains("Perennial", case=False)), "NEW_CLASS_ID"] = "2"
filtered_gdf.loc[(gdf["CLASS_ID"] == 10) & (filtered_gdf["CLASS_NAME"].str.contains("Brush|Shrub", case=False)), "NEW_CLASS_ID"] = "3"

# Verify if any CLASS_ID = 10 still has missing values
print(filtered_gdf[filtered_gdf["CLASS_ID"] == 10][["CLASS_ID", "CLASS_NAME", "NEW_CLASS_ID"]])
```

```python
# Save as a GeoParquet file
filtered_gdf = filtered_gdf[["NEW_CLASS_ID", "geometry"]].rename(columns={"NEW_CLASS_ID": "class_id"})
filtered_gdf["class_id"] = filtered_gdf["class_id"].astype(int)

filtered_gdf.to_parquet("../01_processed_data/land_cover.parquet", index=False)
filtered_gdf
```
