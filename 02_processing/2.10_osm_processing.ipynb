## Solar Sites

```python
import geopandas as gpd
power_features = gpd.read_file("../00_data/power_features.geojson")
```

```python
power_subset[(power_subset['power'] == 'generator') 
             & (power_subset['generator:source'] == 'solar')]
```

```python
# Filter to remove rooftop panels
solar_sites = power_subset[
    (power_subset['power'] == 'generator') &
    (power_subset['generator:source'] == 'solar') &
    ~(  # Negate the rooftop condition
        (power_subset['location'] == 'roof') |
        (power_subset['generator:place'] == 'roof')
    )
]
solar_sites = solar_sites.reset_index()
solar_sites
```

```python
solar_sites['lamp_type'].value_counts()
```

```python
# Exclude 'solar_lamp' while keeping rows where lamp_type is NaN or something else
solar_sites_filtered = solar_sites[~(solar_sites['lamp_type'] == 'solar_lamp')]
```

```python
solar_sites_filtered['generator:output:electricity'].value_counts()
```

```python
import pandas as pd

# Exclude rows with large MW capacity
def is_too_large(capacity):
    if pd.isnull(capacity):
        return False
    if isinstance(capacity, str):
        if 'MW' in capacity:
            try:
                value = float(capacity.split()[0])
                return value >= 1.0
            except:
                return True  # If not parseable, exclude
    return False

solar_sites_filtered = solar_sites_filtered[~solar_sites_filtered['generator:output:electricity'].apply(is_too_large)]
solar_sites_filtered['generator:output:electricity'].value_counts()
```

```python
solar_sites_filtered.to_file("../00_data/solar_sites.geojson", driver="GeoJSON")
```

```python
# Create a copy to avoid modifying original
gdf = solar_sites_filtered.copy()

# Get a representative point inside each polygon
gdf['representative_point'] = gdf.geometry.representative_point()

# Extract latitude and longitude from the representative points
gdf['latitude'] = gdf['representative_point'].y
gdf['longitude'] = gdf['representative_point'].x

# Select only id, latitude, and longitude
csv_df = gdf[['id', 'latitude', 'longitude']].reset_index(drop=True)
csv_df
```

```python
# Save to CSV
csv_df.to_csv('../00_data/solar_sites_coords.csv', index=False)
```
