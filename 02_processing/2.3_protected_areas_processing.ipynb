# Processing Protected Areas Map

```python
import geopandas as gpd
import fiona

# Path to your GDB file
gdb_path = "../00_data/WDPA_WDOECM_Mar2025_Public_PHL/WDPA_WDOECM_Mar2025_Public_PHL.gdb"

# List all available layers in the GDB
layers = fiona.listlayers(gdb_path)
print("Layers:", layers)

# Read a specific layer
gdf = gpd.read_file(gdb_path, layer=layers[0])  # Change layer index if needed

gdf
```

```python
# Filter gdf
filtered_gdf = gdf[["WDPAID", "NAME", "geometry"]]

# Save as a GeoParquet file
filtered_gdf.to_parquet("../01_processed_data/protected_areas.parquet", index=False)
filtered_gdf
```
